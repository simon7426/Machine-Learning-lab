```python
import numpy as np  
import matplotlib.pyplot as plt  
import pandas as pd  
from apyori import apriori 
```

```python
df=pd.read_csv('groceries.csv',header=None)
print(df.shape)
```

```python
count={}
for x in range(0,9835):
    #print(x)
    for y in range(0,32):
        if df.iloc[x,y] in count:
            count[df.iloc[x,y]]+=1
        else:
            count[df.iloc[x,y]]=1
for keys,values in count.items():
    print(keys,values)
```

```python
record = []
for i in range(0,9835):
    record.append([str(df.values[i,j]) for j in range (0,32)])
```

```python
association_rules = apriori(record, min_support=0.0045, min_confidence=0.2, min_lift=3, min_length=2)  
association_results = list(association_rules)  
print(len(association_results)) 
```

```python
dic = {'item0':[],'item1':[],'Support':[],'Confidence':[],'Lift':[]}
```

```python
for item in association_results:
    pair = item[0] 
    items = [x for x in pair]
    dic['item0'].append(items[0])
    dic['item1'].append(items[1])
    dic['Support'].append(str(item[1]))
    dic['Confidence'].append(str(item[2][0][2]))
    dic['Lift'].append(str(item[2][0][3]))
```

```python
df1 = pd.DataFrame.from_dict(dic)
with pd.option_context('display.max_rows',None,'display.Max_column',None):
    display(df1)
```
